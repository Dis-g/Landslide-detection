```python
import sys
import os
import matplotlib.pyplot as plt
import numpy as np
sys.path.append(os.path.abspath('../scripts'))
```

```python
from load_data import tiftopython
```

```python
elevation, slope = tiftopython()
```

```python
# Show elevation
plt.figure(figsize=(8, 6))
plt.title("Elevation Map (m)")
plt.imshow(elevation, cmap='terrain')
plt.colorbar(label='Elevation (m)')
plt.show()
```

```python
# Show slope
plt.figure(figsize=(8, 6))
plt.title("Slope Map (degrees)")
plt.imshow(slope, cmap='viridis')
plt.colorbar(label='Slope ()')
plt.show()
```
